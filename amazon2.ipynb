```python
#cod 2
# Importando as bibliotecas necessárias
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By

# Configurações do webdriver
options = webdriver.FirefoxOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--ignore-ssl-errors=yes')
options.add_argument('--start-maximized')

# Inicializa o navegador
firefox = webdriver.Remote(command_executor="http://localhost:4444/wd/hub", options=options)

# URL da página de avaliações
url = "https://www.amazon.com.br/Unique-Stories-conte%C3%BAdo-fortalecer-oportunidades/product-reviews/8582892136/ref=cm_cr_arp_d_paging_btm_prev_1?ie=UTF8&pageNumber=1&reviewerType=all_reviews"

# Abre a URL
firefox.get(url)

# Espera um tempo para que a página carregue completamente
sleep(5)

# Extrai as informações das avaliações
reviews = []

# Encontra todos os elementos que contêm informações das avaliações
review_elements = firefox.find_elements(By.CSS_SELECTOR, "div.a-section.review.aok-relative")

for review_element in review_elements:
    # Extrai as informações de cada avaliação
    author = review_element.find_element(By.CSS_SELECTOR, "span.a-profile-name").text
    rating = review_element.find_element(By.CSS_SELECTOR, "span.a-icon-alt").get_attribute("innerHTML")
    
    # Título
    title_element = review_element.find_element(By.CSS_SELECTOR, "a.review-title")
    title = title_element.text if title_element else "Sem título"
    
    content = review_element.find_element(By.CSS_SELECTOR, "span.review-text-content span").text
    date = review_element.find_element(By.CSS_SELECTOR, "span.review-date").text
    
    # Verifica se a avaliação é verificada
    verified_element = review_element.find_elements(By.XPATH, ".//span[contains(text(), 'Compra verificada')]")
    verified = "Compra verificada" if verified_element else "Não verificada"
    
    # Tenta encontrar a URL da imagem da avaliação
    try:
        image_url = review_element.find_element(By.CSS_SELECTOR, "img.review-image-tile").get_attribute("src")
    except:
        image_url = None
    
    # Adiciona as informações à lista de avaliações
    reviews.append({
        "author": author,
        "rating": rating,
        "title": title,
        "content": content,
        "date": date,
        "verified": verified,
        "image_url": image_url
    })

# Imprime as avaliações no formato desejado
print("author\trating\ttitle\tcontent\tdate\tverified\timage_url")
for review in reviews:
    print(f"{review['author']}\t{review['rating']}\t{review['title']}\t{review['content']}\t{review['date']}\t{review['verified']}\t{review['image_url']}")

# Fechar o navegador
firefox.quit()

```

```python
#combinação de page_links parte 2, com todos os links txt
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import pandas as pd

options = webdriver.FirefoxOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--ignore-ssl-errors=yes')
options.add_argument('--start-maximized')

# Inicializa o navegador
firefox = webdriver.Remote(command_executor="http://localhost:4444/wd/hub", options=options)

# Lista para armazenar todas as avaliações
all_reviews = []

# URLs das páginas de avaliações
# Abrir o arquivo de texto contendo os links
with open('/caminho/para/o/arquivo/links.txt', 'r') as file:
    # Ler os links do arquivo e armazená-los em uma lista
    urls = file.readlines()

# Iterar sobre os links e processá-los
for url in urls:
    # Remover espaços em branco extras e caracteres inválidos
    url = url.strip()

    # Verificar se o URL é válido
    if url:
        try:
            # Abre a URL
            firefox.get(url)

            # Espera um tempo para que a página carregue completamente
            sleep(10)

            # Extrai as informações das avaliações
            reviews = []

            # Encontra todos os elementos que contêm informações das avaliações
            review_elements = firefox.find_elements(By.CSS_SELECTOR, "div.a-section.review.aok-relative")

            for review_element in review_elements:
                # Extrai as informações de cada avaliação
                author = review_element.find_element(By.CSS_SELECTOR, "span.a-profile-name").text
                rating = review_element.find_element(By.CSS_SELECTOR, "span.a-icon-alt").get_attribute("innerHTML")

                # Título
                try:
                    title_element = review_element.find_element(By.CSS_SELECTOR, "a.review-title span")
                    title = title_element.text if title_element else "Sem título"
                except:
                    title = "Sem título"

                content = review_element.find_element(By.CSS_SELECTOR, "span.review-text-content span").text
                date = review_element.find_element(By.CSS_SELECTOR, "span.review-date").text

                # Verifica se a avaliação é verificada
                verified_element = review_element.find_elements(By.XPATH, ".//span[contains(text(), 'Compra verificada')]")
                verified = "Compra verificada" if verified_element else "Não verificada"

                # Tenta encontrar a URL da imagem da avaliação
                try:
                    image_url = review_element.find_element(By.CSS_SELECTOR, "img.review-image-tile").get_attribute("src")
                except:
                    image_url = None

                # Adiciona as informações à lista de avaliações
                reviews.append({
                    "URL": url,
                    "author": author,
                    "rating": rating,
                    "title": title,
                    "content": content,
                    "date": date,
                    "verified": verified,
                    "image_url": image_url
                })

            # Adiciona as avaliações deste URL à lista geral de avaliações
            all_reviews.extend(reviews)
        except Exception as e:
            print(f"Erro ao processar URL {url}: {e}")
    else:
        print("URL inválido: ", url)

# Fechar o navegador
firefox.quit()

# Cria um DataFrame a partir da lista de avaliações
df = pd.DataFrame(all_reviews)

# Exibe o DataFrame
df

```
